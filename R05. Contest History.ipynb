### Data

```python
history_df = pd.read_csv(r"C:\Users\James\Downloads\draftkings-contest-entry-history (15).csv")
```

### Sample

```python
history_df = history_df[history_df['Sport'] == "MLB"]
history_df = history_df[history_df['Game_Type'] == "Classic"]
```

### Clean

Convert data types

```python
history_df['Winnings_Non_Ticket'] = history_df['Winnings_Non_Ticket'].str.replace("$", "")
history_df['Winnings_Non_Ticket'] = history_df['Winnings_Non_Ticket'].str.replace(",", "").astype(float)
```

```python
history_df['Entry_Fee'] = history_df['Entry_Fee'].str.replace("$", "").astype(float)
```

```python
history_df['Contest_Date_EST'] = pd.to_datetime(history_df['Contest_Date_EST'])
```

Calculate ROI

```python
history_df['ROI'] = (history_df['Winnings_Non_Ticket'].astype(float) - history_df['Entry_Fee'].astype(float))/ history_df['Entry_Fee'].astype(float)
```

Calculate percentilke finish

```python
history_df['Percentile'] = 1 - ((history_df['Place'] - 1) / history_df['Contest_Entries'])
history_df['P95'] = (history_df['Percentile'] > 0.95).astype(int)
history_df['P99'] = (history_df['Percentile'] > 0.99).astype(int)
```

Calculate payout rates

```python
history_df['Paid'] = (history_df['Winnings_Non_Ticket'].astype(float) > 0).astype(int)
history_df['Field_Paid'] = history_df['Places_Paid'] / history_df['Contest_Entries']
```

### Analysis

```python
history_df[(history_df['Contest_Date_EST'] >= '2026-01-01') &
           (history_df['Contest_Date_EST'] <= '2027-01-01')][history_df['Entry_Fee'].astype(float) == 4.00][['Points', 'Winnings_Non_Ticket', 'Entry_Fee', 'ROI', 'Percentile', 'P95', 'P99', 'Paid', 'Field_Paid']].describe()
```
